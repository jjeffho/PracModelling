```python
#PCA
```

```python
#Perform feature scaling 
from sklearn.preprocessing import StandardScaler
```

```python
std_X = StandardScaler().fit_transform(data)
```

```python
#Checking transformed dataset
std_X
```

```python
#Calcuaing covariance matrix
```

```python
mean_vec = np.mean(std_X, axis=0)
cov_mat = (std_X - mean_vec).T.dot((std_X - mean_vec)) / (std_X.shape[0] - 1)

print('Covariance matrix:')
print(cov_mat)
```

```python
#Calculating eigenvalues and eigenvectors
```

```python
eig_vals, eig_vecs = np.linalg.eig(cov_mat)
```

```python
print("Eigenvectors")
print(eig_vecs,'\n')
print('Eigenvectors')
print (eig_vecs)
```

```python
#Printing out PCs
for i in sorted(eig_vals, reverse=True):
    print(round(i,3))

print()    
print ('Sum of all eigenvalues is ' + str(sum(eig_vals)))

PC16 = sum(eig_vals[0:16]) / sum(eig_vals) * 100

print("Top 16 Eigen Values: {:.3f}".format(PC16))
```

```python
#K Means Clustering 
```

```python
from sklearn.cluster import KMeans
```

```python
kmean=KMeans(n_clusters=3)
```

```python
kmean.fit(data)
```

```python
#Shape of centroids
centroids = kmean.cluster_centers_
print("Shape of Centroids Array: " + str(centroids.shape))

print()
print(centroids)
```

```python
#Show percentage the number samples assigned to each cluster as such:
from collections import Counter
```

```python
labels = kmean.labels_
c = Counter(labels)
print(c.most_common())

for cluster_number in range(0,3):
  print("Cluster {} contains {} samples with percentage of {:.2f}%".format(
      cluster_number, c[cluster_number], c[cluster_number]/sum(c.values()) * 100))
```
